```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd  

sns.set()
%matplotlib inline
```

```python
# K-meansによる分類

from sklearn.cluster import  KMeans
from sklearn.datasets import  make_blobs

n_samples=1000
n_centers=4
seed=1
# データセットの生成
blob,blob_labels=make_blobs(n_samples=n_samples,n_features=2,
                            centers=n_centers,random_state=seed)

# データセットをK-meansで分類
clusters_blos=KMeans(n_clusters=n_centers,random_state=seed).fit_predict(blob)

# ランダムデータを分類
uniform=np.random.rand(n_samples,2)
clusters_random=KMeans(n_clusters=n_centers,random_state=seed).fit_predict(uniform)


```

```python
fig=plt.figure()
plt.subplot(221)
plt.scatter(blob[:,0],blob[:,1],c=blob_labels,cmap='gist_rainbow')
plt.title('ランダムに生成した4つのblob')
plt.axis('off')

plt.subplot(222)
plt.scatter(blob[:,0],blob[:,1],c=clusters_blos,cmap='gist_rainbow')
plt.title('KMMで分類したblob')
plt.axis('off')

plt.subplot(223)
plt.scatter(uniform[:,0],uniform[:,1])
plt.title('ランダムに生成した1000の値')
plt.axis('off')


plt.subplot(224)
plt.scatter(uniform[:,0],uniform[:,1],c=clusters_random,cmap='gist_rainbow')
plt.title('ランダムに生成した1000の値をK-meansで分類')
plt.axis('off')
```
