```python
# json形式
obj = """
{"name": "Wes",
 "places_lived": ["United States", "Spain", "Germany"],
 "pet": null,
 "siblings": [{"name": "Scott", "age": 30, "pets": ["Zeus", "Zuko"]},
              {"name": "Katie", "age": 38,
               "pets": ["Sixes", "Stache", "Cisco"]}]
}
"""

import json

result=json.loads(obj)
type(result)

# jsonに再変換
dump=json.dumps(result)
type(dump)

datas=pd.DataFrame(result['siblings'])

# datafameをjson（str)に変換
datas.to_json()
```
