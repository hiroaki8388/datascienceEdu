```python
import numpy as np
import pandas as pd
import scipy as sp
from  scipy import stats

from matplotlib import  pyplot as plt
import seaborn as sns

sns.set()

%matplotlib inline
%precision 3

np.random.seed(1)
```

```python
loc=4
scale=0.8
norm_dist=stats.norm(loc=loc,scale=scale)
```

```python
# t値の分布は微妙に標準正規分布とずれる
samples=[norm_dist.rvs(10) for _ in range (10000)]
t_arr=[
    (sp.mean(samples[i])-loc)/(sp.std(samples[i],ddof=1)/sp.sqrt(10)) for i in range(10000)
]
```

```python
# すそが微妙におおきい(分散が1から微妙にずれてる)
sns.distplot(t_arr)
x=np.arange(-8,8.1,0.1)
plt.plot(x,stats.norm.pdf(x=x),'k--')
```

```python
# 自由度9のt分布に一致
sns.distplot(t_arr)
x=np.arange(-8,8.1,0.1)
plt.plot(x,stats.t.pdf(x=x,df=9),'k--')
```
